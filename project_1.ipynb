```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
import seaborn as sns

```

```python
#load  the Dataset
df=pd.read_csv("/content/Billionaires Statistics Dataset.csv")
df.head()
df.tail()
df.info()
df.describe()


```

```python
#  Calculate list of 10 top_countries billionair
# Get the top 10 countries by number of billionaires
top_countries = df['country'].value_counts().head(10)

# Create a list of 10 different colors (one for each country)
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan', 'yellow']

# Plot the bar chart with different colors for each bar
plt.figure(figsize=(10,6))

# Use top_countries instead of country
bars = plt.bar(top_countries.index, top_countries.values, color=colors)

# Add titles and labels
plt.title('Top 10 Countries by Number of Billionaires')
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45)

# Add a legend with different colors for each country
plt.legend(bars, top_countries.index, title="Countries", loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#                                     # Calculate a List of 10 top countries’ billionaire
 #Using a bar chart in this context is crucial because it effectively illustrates the distribution of billionaires across different countries, making it easy to compare the number of billionaires in each nation.
 #The chart shows that the United States has the most billionaires, followed by China These two countries are home to the richest people, showing their economic power. Other countries like Germany and India also have many billionaires, but most of the world's billionaires live in just a few wealthy countries. This shows that richer countries tend to have more billionaires.
```

```python
 #2 how does net worth  vary by category (industry)?
```

```python

from matplotlib.patches import Patch

# Calculate the average net worth by category
average_net_worth_by_category = df.groupby('category')['finalWorth'].mean()

# Define colors for each category
colors = ['skyblue', 'orange', 'lightgreen', 'purple', 'pink', 'gray', 'brown', 'yellow', 'lightblue', 'darkgreen',
          'cyan', 'magenta', 'lime', 'coral', 'gold', 'violet', 'khaki', 'teal']
colors = colors[:len(average_net_worth_by_category)]

# Create a horizontal bar chart with the custom color list
ax = average_net_worth_by_category.plot(kind='barh', figsize=(12, 6), color=colors)

# Customize the chart
plt.title('Average Net Worth of Billionaires by Industry Category', color='blue')
plt.xlabel('Average Net Worth (in USD)', color='red')
plt.ylabel('Industry Category', color='red')

# Create custom legend using colored patches
legend_patches = [
    Patch(color='skyblue', label='Fashion & Retail'),
    Patch(color='orange', label='Automotive'),
    Patch(color='lightgreen', label='Technology'),
    Patch(color='purple', label='Finance & Investments'),
    Patch(color='pink', label='Media & Entertainment'),
    Patch(color='gray', label='Telecom'),
    Patch(color='brown', label='Diversified'),
    Patch(color='yellow', label='Food & Beverage'),
    Patch(color='lightblue', label='Healthcare'),
    Patch(color='darkgreen', label='Construction & Engineering'),
    Patch(color='cyan', label='Energy'),
    Patch(color='magenta', label='Logistics'),
    Patch(color='lime', label='Sports'),
    Patch(color='coral', label='Real Estate'),
    Patch(color='gold', label='Services'),
    Patch(color='violet', label='Metals & Mining'),
    Patch(color='khaki', label='Manufacturing'),
    Patch(color='teal', label='Gambling & Casino')
]
plt.legend(handles=legend_patches, title='Industry Category', loc='upper right')

plt.tight_layout()
plt.show()
```

```python

from matplotlib.lines import Line2D

# Define age labels and bins
age_labels = ['0-30', '31-50', '51-70', '71+']
age_bins = [0, 30, 50, 70, float('inf')]  # Use float('inf') for the open-ended interval

# Create a new column for age groups
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Set color palette
palette = sns.color_palette("husl", len(age_labels))  # Using 'husl' palette for distinct colors

# Create the violin plot with different colors
plt.figure(figsize=(10, 6))
sns.violinplot(x='age_group', y='finalWorth', data=df, palette=palette)

# Add title and axis labels
plt.title('Distribution of Net Worth of Billionaires by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Net Worth (in USD)')
plt.xticks(rotation=45)

# Create custom legend entries without a loop
legend_entry1 = Line2D([0], [0], marker='o', color='w', label=f'0-30 ({len(df[df["age_group"] == "0-30"])} Billionaires)',
                       markerfacecolor=palette[0], markersize=10)
legend_entry2 = Line2D([0], [0], marker='o', color='w', label=f'31-50 ({len(df[df["age_group"] == "31-50"])} Billionaires)',
                       markerfacecolor=palette[1], markersize=10)
legend_entry3 = Line2D([0], [0], marker='o', color='w', label=f'51-70 ({len(df[df["age_group"] == "51-70"])} Billionaires)',
                       markerfacecolor=palette[2], markersize=10)
legend_entry4 = Line2D([0], [0], marker='o', color='w', label=f'71+ ({len(df[df["age_group"] == "71+"])} Billionaires)',
                       markerfacecolor=palette[3], markersize=10)

# Add custom legend to the plot
plt.legend(handles=[legend_entry1, legend_entry2, legend_entry3, legend_entry4], title='Age Group', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# violin Plot
#3 The distribution of net worth by age group
#It shows the distribution of data across different age groups
#The chart indicates that the age group 51-70 has the highest number of billionaires, followed by the 71+ and 31-50 age groups. However, the age group with the fewest billionaires is 0-30
```

```python

## 4  correlation between age and net worth of billionaires

# Create Scatter Plot with Regression Line
plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='finalWorth', data=df, logx=True, scatter_kws={'s': 50})
plt.title('Scatter Plot of Age vs. Net Worth of Billionaires with Regression Line')
plt.xlabel('Age')
plt.ylabel('Net Worth (in USD)')
plt.xscale('linear')
plt.yscale('log')  # Use log scale for better visualization of wealth distribution
plt.tight_layout()
plt.show()
```

```python
# title Default title text
# Scatter plot with regression line :
#4  correlation between age and net worth of a billionaire
# It helps to visualize the relationship between the age and net worth of billionaires.
# Our analysis shows a moderate positive correlation between age and net worth among billionaires, indicating that older individuals tend to have higher wealth due to accumulated experience and resources. However, several young billionaires also exhibit significant net worth, suggesting that wealth accumulation can occur at any age. This highlights the influence of industry and individual circumstances in determining financial success.

```
