```python
import matplotlib.pyplot as plt
import numpy as np

print(plt.style.available)
print(plt.style.use('seaborn-whitegrid'))
plt.style.use('ggplot')
alternative_style = 'ggplot'
```

```python
##1
#Create a plot of the quadratic function
x = np.linspace(-2, 4, 100)
y = x**2-2*x+1
plt.plot(x, y)
plt.show()
```

```python
##2 Basic line plot
```

```python
x = np.linspace(-2, 2, 100)
f_x = x**3
g_x = np.exp(x)
x_log = np.linspace(0.1, 2, 100)  # Avoid log(0)
h_x = np.log(x_log)

plt.plot(x, f_x, label="f(x) = x^3", color='blue')
plt.plot(x, g_x, label="g(x) = e^x", color='green')
plt.plot(x_log, h_x, label="h(x) = log(x)", color='red')
plt.legend()
plt.title("Multiple Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
##3
# Create a plot with the following functions, each with a different color and line style
```

```python
x = np.linspace(-2, 2, 100)
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5

plt.plot(x, y1, 'r-', label='x^2')  # Red solid line
plt.plot(x, y2, 'g--', label='x^3')  # Green dashed line
plt.plot(x, y3, 'b-.', label='x^4')  # Blue dash-dot line
plt.plot(x, y4, 'm:', label='x^5')  # Magenta dotted line
plt.legend()
plt.title('Styled Lines')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5

plt.plot(x, y1, 'r-', label='x^2')  # Red solid line
plt.plot(x, y2, 'g--', label='x^3')  # Green dashed line
plt.plot(x, y3, 'b-.', label='x^4')  # Blue dash-dot line
plt.plot(x, y4, 'm:', label='x^5')  # Magenta dotted line
plt.legend()
plt.title('Styled Lines')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Axes Limits and Scaling

x = np.linspace(-10, 10, 400)
y = 1 / (x**2 + 1)

plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(0, 1)
plt.title('Function with Custom Limits')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Reversed axes
plt.plot(x, y)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title('Function with Reversed Axes')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Tight and equal axis
plt.plot(x, y)
plt.axis('tight')
plt.title('Tight Axis')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

plt.plot(x, y)
plt.axis('equal')
plt.title('Equal Axis')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# 7 Basic scatter plot
x = np.random.rand(65)
y = np.random.rand(65)

# Using plt.plot()
plt.plot(x, y, 'o')
plt.title('Scatter Plot with plt.plot()')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Using plt.scatter()
plt.scatter(x, y)
plt.title('Scatter Plot with plt.scatter()')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
##8 Customizing Scatter Plots
x = np.linspace(0, 4*np.pi, 400)
y = x * np.sin(x)

plt.scatter(x, y)
plt.plot(x, y, 'r-')
plt.title('Custom Scatter Plot')
plt.xlabel('x')
plt.ylabel('x sin(x)')
plt.show()
```

```python
##9  Combined plot
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
f_x = x * np.cos(x)

plt.plot(x, f_x, label="f(x) = x cos(x)", color='blue')
scatter_x = np.arange(-2 * np.pi, 2 * np.pi, np.pi / 4)
plt.scatter(scatter_x, scatter_x * np.cos(scatter_x), color='red', label='Scatter points')
plt.legend()
plt.title("Combined Line and Scatter Plot")
plt.show()
```

```python
#  10 Custom plot with alternative Style
plt.style.use(alternative_style)

x = np.linspace(-10, 10, 200)
f_x = np.tan(x)
g_x = x * np.exp(-x**2)
h_x = np.sin(x) / x

plt.plot(x, f_x, label="tan(x)", linestyle='--')
plt.plot(x, g_x, label="x * exp(-x^2)", linestyle='-.')
plt.plot(x, h_x, label="sin(x)/x", linestyle=':')

plt.title("Custom Plot with Alternative Style")
plt.legend()
plt.show()
```

```python
#  11 style comparison
# dark_background style
plt.style.use('dark_background')
plt.plot(x, f_x, 'r-', label="f(x) = x^2")
plt.show()

# Seaborn-whitegrid style
plt.style.use('seaborn-whitegrid')
plt.plot(x, f_x, 'r-', label="f(x) = x^2")
plt.show()

# Alternative style (ggplot)
plt.style.use(alternative_style)
plt.plot(x, f_x, 'r-', label="f(x) = x^2")
plt.show()
```
