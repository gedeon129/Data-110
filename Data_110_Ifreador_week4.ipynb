```python
     # Task 1
import matplotlib.pyplot as plt

# Data for offenses and their rates per 100,000 people in 2023
offenses = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (Revised Definition)', 'Robbery']
rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(offenses, rates_2023, color='skyblue')

# Add title and labels
plt.title('Crime Rates per 100,000 People in 2023', fontsize=16)
plt.xlabel('Types of Offenses', fontsize=12)
plt.ylabel('Offenses per 100,000 People', fontsize=12)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=50, ha='right')



# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#  Task 3

import matplotlib.pyplot as plt
import numpy as np

# Offense types
crimes = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']

# Percentage change from 2022 to 2023
crime_rates_2023 = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Calculate the crime rates for 2022:
crime_rates_2022= [ 267.4, 258.8, 1352.1,306.7, 17.7,47.8,67.3]


# Bar width and x indices
bar_width = 0.25
x_indexes = np.arange(len(crimes))

# sample data

# Create the combined bar chart
plt.figure(figsize=(12, 7))
bars_2022 = plt.bar(x_indexes, crime_rates_2022, width=bar_width, label='2022', color='red')
bars_2023 = plt.bar(x_indexes + bar_width, crime_rates_2023, width=bar_width, label='2023', color='green')


# Add labels and title
plt.xlabel('Type of Offense')
plt.ylabel('Number of Offenses')
plt.title('Comparison of Crime Rates: 2022 vs 2023')

plt.xticks(ticks=x_indexes, labels= crimes)


# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()



```
