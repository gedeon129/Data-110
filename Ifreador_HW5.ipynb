```python
# task 3 Fare and Survuval
 ##$ Create a KDE plot to compare the fare distributions for survivors and non-survivors.
 ## Question: Does fare seem to have influenced survival? What do the distributions suggest about socioeconomic status and survival?


 # Create a KDE plot for fare distributions
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic[titanic['survived'] == 1], x='fare', label='survived', fill=True)
sns.kdeplot(data=titanic[titanic['survived'] == 0], x='fare', label='Did Not survive', fill=True)

# Add titles and labels
plt.title('Fare Distribution for Survivors and Non-Survivors')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.xlim(0, 100)
plt.legend()
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))

sns.kdeplot(data=titanic[titanic['pclass'] == 1], x='fare', label='survived', fill=True)
sns.kdeplot(data=titanic[titanic['pclass'] == 0], x='fare', label='Did Not survive', fill=True)
plt.hist(titanic['fare'], bins=15, color='green', edgecolor='white')
plt.xlabel('fare')
plt.ylabel('count')
plt.title('pclass')
plt.grid(axis='x')
plt.xticks([0, 20,40,60])
plt.tick_params(axis='x', color='red')
plt.show()
```

```python


# Plot histograms and KDE plots to compare fare distributions across different passenger classes
plt.figure(figsize=(10, 6))

# Histogram of Fare distribution by Passenger Class (Pclass)
plt.subplot(2, 2, 1)
sns.histplot(data=titanic, x='fare', hue='pclass', multiple='stack', bins=30)
plt.title('Histogram of Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Frequency')

# KDE plot of Fare distribution by Passenger Class
plt.subplot(2, 2, 2)
sns.kdeplot(data= titanic, x='fare', hue='pclass', fill=True)
plt.title('KDE Plot of Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Density')

# Histogram of Survival Rates by Passenger Class
plt.subplot(2, 2, 3)
sns.histplot(data= titanic, x='fare', hue='survived', multiple='stack', bins=30, palette=['red', 'green'])
plt.title('Histogram of Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Frequency')

# KDE plot of Survival Rates by Passenger Class
plt.subplot(2, 2, 4)
sns.kdeplot(data= titanic[titanic['survived'] == 1], x='fare', hue='pclass', fill=True)
plt.title('KDE Plot of Fare Distribution of Survivors by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
#Passengers in first class paid the highest fares, while those in third class paid the least.
#This shows the wealthier passengers had more luxurious accommodations. First-class passengers also had better survival rates because they had better access to lifeboats and received preferential treatment.
#This highlights the clear divisions based on wealth and class on the Titanic, with richer passengers having better chances of survival.
```

```python

!pip install  matplotlib pandas seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# View the first few rows of the dataset
titanic.head()

```

```python
# Create a figure
fig, ax = plt.subplots(figsize=(10, 8))

# Assuming 'titanic' is your DataFrame loaded with the Titanic dataset
# Extract age and survival information for male and female passengers
# Use bitwise '&' instead of 'and' to combine boolean conditions
male_data = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 1)]
female_data = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)]

# Extract ages for male and female survivors
ages = sorted(titanic['age'].unique())  # Get unique ages and sort them
male_counts = male_data['age'].value_counts().sort_index()
female_counts = female_data['age'].value_counts().sort_index()


# Plot males on the left side (negative direction)
ax.barh(ages, -pd.Series(male_counts).reindex(ages).fillna(0), color='blue', label='Male (Survived)')

# Plot females on the right side (positive direction)
ax.barh(ages, pd.Series(female_counts).reindex(ages).fillna(0), color='pink', label='Female (Survived)')

# Set the labels
ax.set_xlabel('Survived Count')
ax.set_ylabel('Age')
ax.set_title('Age and Survival by Sex')

# Add legend
ax.legend()

# Display the plot
plt.show()
```

```python
##The chart likely shows that women had higher survival rates than men across most age groups. This reflects the "women first" policy during the Titanic's evacuation, where females were prioritized for lifeboats. Men, especially adults, had much lower survival rates, as they were less likely to be rescued. The disparity between male and female survival highlights the gender-based rescue practices during the disaster.
```
